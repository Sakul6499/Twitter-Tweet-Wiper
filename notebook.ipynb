```python
f = open(file, 'r', errors = 'ignore')
lines = f.readlines()
```

```python
new_lines = []
for line in lines:
    new_lines.append(line.strip())

lines = new_lines
```

```python
ids = []
for line in lines:
    if line.startswith("\"id\"") or line.startswith("\"id_str\""):
        ids.append(line.split(":")[1].strip().replace("\"", "").replace(",", ""))
```

```python
auth = tweepy.OAuthHandler(api_key, api_secret)
auth_url = auth.get_authorization_url()
print(auth_url)

verify_code = input(
    "Authenticate at %s and then enter you verification code here > " % auth_url
)
auth.get_access_token(verify_code)

client = tweepy.API(auth)
```

```python
for id in ids:
    try:
        print(id + " " + type(id).__name__)
        client.destroy_status(id)
    except Exception as e:
        print("failed (" + str(e) + ")")
    print()
```
